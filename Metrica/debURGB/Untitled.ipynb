```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
def printImgs(images, n=1, m=1, title=''):
    
    images = images[:n*m]
    fig, axs = plt.subplots(
        nrows=n, ncols=m, 
        figsize=(15, 20), 
        subplot_kw={
            'xticks': [], 
            'yticks': []
        }
    )
    for ax, image in zip(axs.flat, images):
        ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.savefig(title)    
    plt.show()
```

```python
lsImg = []

for k in range(1,25):#25
    #nameimg = 'kodim' + str(k) + '.png'
    nameimg = 'kodim' + str(k) + '_small.png'
    
    # Load clean image
    ruta = './deblurring/'
    fname = ruta + nameimg
    imgOut = cv2.imread(fname)
    lsImg.append(imgOut)
```

```python
lsImg = np.array(lsImg)
```

```python
ind = [3,8,9,16,17,18]
```

```python
imgG = np.array(ind)
lsimgG = lsImg[imgG]
```

```python
printImgs(lsimgG, 1, len(lsimgG),fname)
```

```python
lsImgCh = np.delete(lsImg, ind)
```

```python
printImgs(lsImgCh, 3, 6)
```

```python
lsImg = np.concatenate((lsImgCh,lsimgG))
```

```python
fname = 'KodimColorTotal.png'
printImgs(lsImg, 4, 6,fname)
```
